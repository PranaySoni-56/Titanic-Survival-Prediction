##### Handle Outliers

```python
# Check percentage of outlier
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Age'] < lower_bound) | (df['Age'] > upper_bound)]
print(f"Total outliers in Age: {len(outliers)}")
```

```python
# Replace outlier with median
df.loc[df['Age'] > upper_bound, 'Age'] = df['Age'].median()
```

### Feature Engineering

```python
#Create a new Feature of Family size
# include passenger
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

```python
# Create a new feature isalone
df['IsAlone'] = (df['FamilySize'] == 1).astype(int)
```

```python
# Extract title from name
df['Title'] = df['Name'].str.extract(r' ([A-Za-z]+)\.', expand=False) 
# Group rare titles
df['Title'] = df['Title'].replace(['Lady', 'Countess', 'Capt', 'Col', 'Don', 'Dr', 
                                   'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
df['Title'] = df['Title'].replace(['Mlle', 'Ms'], 'Miss')
df['Title'] = df['Title'].replace('Mme', 'Mrs')
```

```python
# Convert Cabin information to a cabin
df['Has_cabin'] = df['Cabin'].notna().astype(int)
```

```python
# Encode Categorical Feature
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})  # Binary encoding
df = pd.get_dummies(df, columns=['Embarked', 'Title'], drop_first=True)  # One-hot encoding
```

```python
# Bucket Fare and age
df['FareBin'] = pd.qcut(df['Fare'], 4, labels=[1, 2, 3, 4])  # Quartiles
df['AgeBin'] = pd.cut(df['Age'], bins=[0, 12, 20, 40, 60, 100], labels=[1, 2, 3, 4, 5])
```

### Machine Learning Using Classification Model

```python
'''
Steps to Apply ML Models:--
 --> Split data into Train and test set
 --> Scale Numerical Feature
 --> Training Multiple Models and compare performance
 --> Evaluate model using Accuracy '''
```

```python
from sklearn.model_selection import train_test_split

# Select features and target variables
X = df.drop(columns=['Survived','PassengerId','Name','Ticket'])
# Target Variable
Y = df['Survived']  

# Train Test split( 80% and 20% ratio)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
''' Now We are Trained Following Models :-
Decision Tree
Random Forest
XG Boost
Naive Bayes
'''
```

```python
X_train.dtypes
```

```python
# Drop the cabin column
X_train = X_train.drop(columns=['Cabin'])
X_test = X_test.drop(columns=['Cabin'])
```

```python
# Convert 'FareBin' & 'AgeBin' categorical to numerical datatype

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

X_train['FareBin'] = le.fit_transform(X_train['FareBin'])
X_test['FareBin'] = le.transform(X_test['FareBin'])

X_train['AgeBin'] = le.fit_transform(X_train['AgeBin'])
X_test['AgeBin'] = le.transform(X_test['AgeBin'])
```

```python
X_train.dtypes
```

```python
# Decision Tree Classifier

from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, Y_train)

# Predict
Y_pred = dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc_scr = accuracy_score(Y_test, Y_pred)
acc_scr
```

```python
# Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train,Y_train)
Y_predict = rf.predict(X_test)
```

```python
acc = accuracy_score(Y_test, Y_predict)
acc
```

### Building a Predictive System

```python
# input data
input_data = (3,1,22.0,1,0,7.25,2,0,0,1,0,0,0,1,0,0,1,2)

# convert input data to numpy array
input_data_as_numpy = np.asarray(input_data)

# Reshape the array
input_data_reshape = input_data_as_numpy.reshape(1,-1)

# Make Prediction
prediction = rf.predict(input_data_reshape)

if prediction[0] == 0:
    print("The Person is unlikely to survive.")
else:
    print("The Person is likely to survive.")
```
